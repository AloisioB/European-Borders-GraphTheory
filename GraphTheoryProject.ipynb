```python
import networkx as nx

G = nx.Graph()
G.add_node('Romania', militar_alliance = "NATO")
G.add_node('Spain', militar_alliance = "NATO")
G.add_node('Finland', militar_alliance = "NATO")
G.add_node('Latvia', militar_alliance = "NATO")
G.add_node('San_Marino', militar_alliance = "NATO")
G.add_node('Portugal', militar_alliance = "NATO")
G.add_node('Norway', militar_alliance = "NONE")
G.add_node('Slovenia', militar_alliance = "NATO")
G.add_node('Italy', militar_alliance = "NATO")
G.add_node('Switzerland', militar_alliance = "NATO")
G.add_node('Czech_Republic', militar_alliance = "NATO")
G.add_node('Moldova', militar_alliance = "NATO")
G.add_node('Albania', militar_alliance = "NATO")
G.add_node('Luxembourg', militar_alliance = "NATO")
G.add_node('Sweden', militar_alliance = "NATO")
G.add_node('Austria', militar_alliance= "NATO")
G.add_node('Denmark', militar_alliance = "NATO")
G.add_node('Hungary', militar_alliance = "NATO")
G.add_node('Germany', militar_alliance = "NATO")
G.add_node('Russia', militar_alliance = "NATO")
G.add_node('Lithuania', militar_alliance = "NATO")
G.add_node('Netherlands', militar_alliance = "NATO")
G.add_node('Georgia', militar_alliance = "NONE")
G.add_node('North_Macedonia', militar_alliance = "NATO")
G.add_node('Estonia', militar_alliance = "NATO")
G.add_node('Ukraine', militar_alliance = "NATO")
G.add_node('Bosnia', militar_alliance = "NATO")
G.add_node('Greece', militar_alliance = "NATO")
G.add_node('Croatia', militar_alliance = "NATO")
G.add_node('Ireland', militar_alliance = "NATO")
G.add_node('Liechtenstein', militar_alliance = "NATO")
G.add_node('Slovakia', militar_alliance = "NATO")
G.add_node('Bulgaria', militar_alliance = "NATO")
G.add_node('France', militar_alliance = "NATO")
G.add_node('Kosovo', militar_alliance = "NATO")
G.add_node('United_Kingdom', militar_alliance = "NATO")
G.add_node('Armenia', militar_alliance = "NATO")
G.add_node('Belarus', militar_alliance = "NATO")
G.add_node('Montenegro', militar_alliance = "NATO")
G.add_node('Serbia', militar_alliance = "NATO")
G.add_node('Turkey', militar_alliance = "NATO")
G.add_node('Belgium', militar_alliance = "NATO")

G.add_edge("Portugal", "Spain", weight=6)
G.add_edge("Spain", "France", weight=1)
G.add_edge("France", "Italy", weight=8)
G.add_edge("France", "Belgium", weight=0)
G.add_edge("France", "Luxembourg", weight=4)
G.add_edge("France", "Germany", weight=0)
G.add_edge("France", "Switzerland", weight=6)
G.add_edge("Belgium", "Luxembourg", weight=6)
G.add_edge("Belgium", "Germany", weight=1)
G.add_edge("Belgium", "Netherlands", weight=3)
G.add_edge("Netherlands", "Germany", weight=0)
G.add_edge("Denmark", "Germany", weight=2)
G.add_edge("Denmark", "Sweden", weight=6)
G.add_edge("Sweden", "Finland", weight=7)
G.add_edge("Sweden", "Norway", weight=0)
G.add_edge("Norway", "Finland", weight=5)
G.add_edge("Norway", "Russia", weight=0)
G.add_edge("Finland", "Russia", weight=4)
G.add_edge("Russia", "Estonia", weight=4)
G.add_edge("Russia", "Latvia", weight=1)
G.add_edge("Russia", "Lithuania", weight=2)
G.add_edge("Russia", "Poland", weight=8)
G.add_edge("Russia", "Belarus", weight=1)
G.add_edge("Russia", "Ukraine", weight=0)
G.add_edge("Russia", "Georgia", weight=9)
G.add_edge("Latvia", "Estonia", weight=3)
G.add_edge("Latvia", "Belarus", weight=3)
G.add_edge("Latvia", "Lithuania", weight=1)
G.add_edge("Belarus", "Ukraine", weight=8)
G.add_edge("Belarus", "Poland", weight=2)
G.add_edge("Belarus", "Lithuania", weight=2)
G.add_edge("Lithuania", "Poland", weight=1)
G.add_edge("Poland", "Germany", weight=5)
G.add_edge("Poland", "Czech_Republic", weight=7)
G.add_edge("Poland", "Slovakia", weight=9)
G.add_edge("Poland", "Ukraine", weight=8)
G.add_edge("Italy", "Switzerland", weight=4)
G.add_edge("Italy", "Austria", weight=0)
G.add_edge("Italy", "Slovenia", weight=6)
G.add_edge("Italy", "San_Marino", weight=7)
G.add_edge("Germany", "Czech_Republic", weight=1)
G.add_edge("Germany", "Austria", weight=7)
G.add_edge("Germany", "Switzerland", weight=3)
G.add_edge("Germany", "Luxembourg", weight=8)
G.add_edge("Austria", "Czech_Republic", weight=9)
G.add_edge("Austria", "Slovakia", weight=1)
G.add_edge("Austria", "Hungary", weight=2)
G.add_edge("Austria", "Slovenia", weight=2)
G.add_edge("Austria", "Switzerland", weight=8)
G.add_edge("Austria", "Liechtenstein", weight=0)
G.add_edge("United_Kingdom", "Ireland", weight=7)
G.add_edge("Ukraine", "Slovakia", weight=0)
G.add_edge("Ukraine", "Hungary", weight=7)
G.add_edge("Ukraine", "Romania", weight=1)
G.add_edge("Ukraine", "Moldova", weight=3)
G.add_edge("Switzerland", "Liechtenstein", weight=9)
G.add_edge("Turkey", "Greece", weight=0)
G.add_edge("Turkey", "Bulgaria", weight=1)
G.add_edge("Turkey", "Georgia", weight=2)
G.add_edge("Turkey", "Armenia", weight=8)
G.add_edge("Romania", "Moldova", weight=3)
G.add_edge("Romania", "Bulgaria", weight=3)
G.add_edge("Romania", "Serbia", weight=1)
G.add_edge("Greece", "Albania", weight=1)
G.add_edge("Greece", "North_Macedonia", weight=9)
G.add_edge("Greece", "Bulgaria", weight=8)
G.add_edge("Armenia", "Georgia", weight=9)
G.add_edge("Serbia", "Hungary", weight=9)
G.add_edge("Serbia", "Bulgaria", weight=1)
G.add_edge("Serbia", "North_Macedonia", weight=9)
G.add_edge("Serbia", "Bosnia", weight=1)
G.add_edge("Albania", "Montenegro", weight=0)
G.add_edge("Albania", "Kosovo", weight=4)
G.add_edge("Albania", "North_Macedonia", weight=4)
G.add_edge("Hungary", "Slovakia", weight=7)
G.add_edge("Hungary", "Romania", weight=5)
G.add_edge("Hungary", "Croatia", weight=6)
G.add_edge("Hungary", "Slovenia", weight=8)
G.add_edge("Bosnia", "Croatia", weight=6)
G.add_edge("Bosnia", "Montenegro", weight=9)
G.add_edge("Slovenia", "Croatia", weight=3)
G.add_edge("Bulgaria", "North_Macedonia", weight=7)
G.add_edge("Croatia", "Serbia", weight=5)
G.add_edge("Croatia", "Montenegro", weight=2)
G.add_edge("Czech_Republic", "Slovakia", weight=9)
G.add_edge("Kosovo", "Serbia", weight=7)
G.add_edge("Kosovo", "Montenegro", weight=8)
G.add_edge("Kosovo", "North_Macedonia", weight=5)
G.add_edge("Montenegro", "Serbia", weight=4)


def Dijkstra(graph, start, end, avoiding_nodes=None, requirement=None):
    if avoiding_nodes is None:
        avoiding_nodes = []

    visited = set()
    previous = {}
    distances = {node: float('inf') for node in graph.nodes()}
    distances[start] = 0

    if requirement is not None:
        if end in avoiding_nodes or graph.nodes[end]['militar_alliance'] != requirement:
            return str("There's no path available!")

    while visited != graph.nodes():
        current_node = min((set(distances.keys()) - visited), key=distances.get)
        if current_node == end:
            path = []
            if requirement is None:
                while current_node != start and current_node not in avoiding_nodes:
                    path.insert(0, current_node)
                    try:
                        current_node = previous[current_node]
                    except KeyError:
                        return str("There's no path available!")
            else:
                while current_node != start and current_node not in avoiding_nodes and graph.nodes[current_node]['militar_alliance'] == requirement:
                    path.insert(0, current_node)
                    try:
                        current_node = previous[current_node]
                    except KeyError:
                        return str("There's no path available!")
            path.insert(0, start)
            return path

        visited.add(current_node)

        for neighbor in graph.neighbors(current_node):
            if requirement is not None:
                if neighbor in avoiding_nodes:
                    continue
                if graph.nodes[neighbor]['militar_alliance'] != requirement:
                    continue
            else:
                if neighbor in avoiding_nodes:
                    continue

            distance = distances[current_node] + graph[current_node][neighbor]['weight']
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous[neighbor] = current_node

    return str("There's no path available!")

start_node = "Portugal"
end_node = "Russia"
avoiding_nodes = ["Austria", "Poland", "Slovenia"]
requirement = "NATO"

shortest_path = Dijkstra(G, start_node, end_node, avoiding_nodes, requirement)
print(shortest_path)
```
